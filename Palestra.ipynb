## Streaming API

```python
from tweepy import Stream
from tweepy.streaming import StreamListener

from tweepy import OAuthHandler
 
consumer_key = "XGS8Xv1y0pWZXyXSHu6c1AVvm"
consumer_secret = "BtzEcusIP67kLJRN7E8xONDwjYwEyASlTXTGxz1jlSbfPkLPes"
access_token = "843892947909332998-SH4Isk6tITrF3Lvku5aNMrP2xWQswrB"
access_secret = "OwmlEjXOcJ8hjauzikRmAYflBInJVYJNcy11MzDYdPwti"

auth = OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)
 

class MyListener(StreamListener):
 
    def on_data(self, data):
        try:
            with open('python.json', 'a') as f:
                f.write(data)
                return True
        except BaseException as e:
            print("Error on_data: %s" % str(e))
        return True
 
    def on_error(self, status):
        print(status)
        return True
 
twitter_stream = Stream(auth, MyListener())
twitter_stream.filter(track=['#PUGPB'])
```
